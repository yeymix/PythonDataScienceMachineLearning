# Añadir Estilo a tus Gráficos

En esta lección quiero enseñarte un recurso de Matplotlib que aplica para cualquier tipo de gráficos que estemos creando. Me refiero a la función `style.use()`, que se utiliza para establecer un **estilo visual** específico para tus gráficos.

Esto es útil porque te permite cambiar rápidamente la apariencia de tus gráficos **sin tener que modificar cada elemento individualmente**. Piénsalo como cambiar de ropa a tus gráficos.

Vamos a tomar de ejemplo esta figura que tiene **dos gráficos** que ya he desplegado para hacer un par de pruebas.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 35, 26, 33]
    })
df
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

Observa que en esos gráficos yo no he modificado nada del estilo visual, solamente he cargado los datos básicos, y mis gráficos tienen **el estilo que viene por defecto** en matplotlib.

Hasta el momento hemos aplicado estilo manualmente a nuestros gráficos, modificando una serie de parámetros (el color de las líneas o del fondo, el grosor o tamaño de los elementos, etc.), pero lo que vamos a hacer gracias a `style,use()` es asignarle a nuestro cuaderno un estilo visual predefinido, para que se aplique a todos los gráficos que genere de ahora en más. Por esa razón, solo va a afectar a los gráficos que renderice de ahora en adelante.

Progemos con us estilo llamado **ggplot**.

```python
plt.style.use('ggplot')

fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

Este es uno de los tantos estilos que Matplotlib tiene preconfigurados.

Si quieres conocer la **lista completa de estilos**, usamos el siguiente comando:

```python
print(plt.style.available)
```

A modo de ejemplo, voy a replicar los mismos gráficos, pero probando algunos de estos estilos:

```python
plt.style.use('seaborn-v0_8-pastel')

fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

```python
plt.style.use('fivethirtyeight')

fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

```python
plt.style.use('dark_background')

fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

Permíteme darte 3 consejos adicionales:
- **Experimenta con diferentes estilos**: Algunos estilos funcionan mejor para ciertos tipos de datos o presentaciones, entonces no dudes en probar varios.
- **Continúa personalizando tus gráficos**: Aunque los estilos proporcionan una base, siempre puedes personalizar aspectos específicos del gráfico según tus necesidades usando los parámetros que ofrece cada tipo de gráfico.
- **Revisa la documentación**: Siempre es útil consultar la documentación oficial de Matplotlib para obtener más detalles y ejemplos.

¡Y eso es todo sobre `style.use()`! Con esta herramienta, puedes hacer que tus gráficos en Python no solo sean informativos sino también visualmente atractivos.
