# Visualizaciones de Datos con Matplotlib

Matplotlib es otra biblioteca creada para Python. Así como **Numpy** fue creada para *cálculos*, y **Pandas** para *estructurar datos*,  **Matplotlib** es un biblioteca especial creada para la *visualización de datos*.

En esta lección te voy a mostrar una pequeña aplicación práctica de Matplotlib, que creo que te va a dar una idea del poder que estás por alcanzar.

Primero tenemos que **importar a Matplotlib**, lo cual se hace de un modo muy similar a como lo venimos haciendo.

```python
import matplotlib.pyplot as plt
```

En esta primera lección vamos a aprender dos métodos básicos de matplotlib que son `plot()` y `show()`:
	- `plot()` sirve para **crear un gráfico**
	- `show()` sirve para **mostrar el gráfico**

Ahora vamos a crear un pequeño conjunto de datos, y luego vamos a mostrarlos en un gráfico de Matplotlib.

```python
numeros = [4, 2, 7, 6, 3]
```

Usaremos `plot()` para crear un gráfico de líneas que represente a `numeros`, y luego a `show()` para mostrar ese gráfico.

```python
plt.plot(numeros)
plt.show()
```

Así de simple, ya tenemos nuestro primer gráfico, que representa a cada número de la lista `numeros`, ubicado en el índice que le corresponde.

Esto se ha visto realmente simple, pero quiero que tengas en cuenta que el comando `plt.show()` hace muchas cosas detrás de la cortina, ya que es el encargado de interactuar con los aspectos gráficos de tu sistema. Este es un comando que debemos usar solamente una vez por sesión de python, y normalmente siempre lo verás al final del código, cuando ya has hecho todas las configuraciones de tu gráfico.

##### Nota especial
Hay una alternativa a `show()`, que consiste en agregar una linea estra en la celda de importaciones, que es el comando:

``%matplotlib inline``

Si agregas esa línea ya no necesitarás escribir `plt.show()` en tu código, sino simplemente agregar un punto y coma (**;**) al final de la última línea del código dedicado a configurar tus visualizaciones. La explicación detallada de por qué y cómo usamos este comando, se encuentra en [este cuaderno adicional](Por%20Qué%20Usamos%20'matplotlib%20inline'.ipynb) que he escrito para ti. De todos modos, la explicación sencilla es: la usamos para que nuestros gráficos puedan renderizarse sin problemas en Jupyter.

### Guardar nuestros gráficos como imágenes independientes

Un detalle adicional, es que puedes guardar tus gráficos como imágenes independientes, usando el método `savefig()`.

```python
plt.plot(numeros)
plt.savefig("mi_grafico.png")
plt.show()
```

Y ahora puedes encontrar el archivo png con tu imagen, en el mismo lugar donde tienes alojado tu cuaderno. Recuerda que también puedes pasarle una ruta completa si quieres alojarla en otra parte de tu ordenador.
