## Matplotlib

Biblioteca especial creada para la visualización de datos.

plot --> crear un gráfico
show --> mostrar un gráfico
Se debe usar sólo una vez por sesión al final.


```python
import matplotlib.pyplot as plt 
# si escribimos la siguiente línea, no será necesario usar el método, con sólo un ; se mosrtará el gráfico:
# `%matplotlib inline` is a magic command in Jupyter notebooks that allows plots generated by matplotlib
#  to be displayed directly in the notebook, rather than in a separate window. 
# This command is used to show the plots inline within the notebook when using matplotlib for data visualization.
%matplotlib inline
```

```python
# crear lista de números
numeros = [4,2,7,6,3]
```

```python
# The code `plt.plot(numeros)` is plotting the values in the list or array `numeros` 
# on a graph using the `plot` function from the `matplotlib.pyplot` library.
plt.plot(numeros);
# plt.show() # Muetra los números ubicados en el índice que le corresponde
plt.savefig('output.png') # Guarda la imagen en un archivo llamado output.png
```

```python
# The code `plt.figure(facecolor="purple")` is creating a new figure for plotting with a purple background color. 
# This sets the background color of the figure to purple before any plotting is done on it.
plt.figure(facecolor="purple")
plt.plot(numeros);
# Tenemos un elemento llamado Acces que incluye todos los elementos que componen el gráfico
```

```python
# Aquí vamos a generar un figure que tiene cuatro Acces
plt.subplots(ncols=2, nrows=2, figsize=(15, 10))
# Prácticamente es un areglo con 4 axes
# (<Figure size 1500x1000 with 4 Axes>,#
# array([[<Axes: >, <Axes: >],
#        [<Axes: >, <Axes: >]], dtype=object))
```

# Introducción a `matplotlib`

## ¿Qué es `matplotlib`?

- Una biblioteca de trazado 2D para Python.
- Permite crear gráficos estáticos, animados e interactivos.

## ¿Por qué usar `matplotlib`?

- Amplia funcionalidad y flexibilidad.
- Gran comunidad y documentación.
- Integración con `NumPy`, `pandas`, y otras bibliotecas de ciencia de datos.

```python
# Crear la pared y el marco
# pared  - Contenedor de cuadros
import matplotlib.pyplot as plt
fig = plt.figure()
#marco - Contenedor de elementos que tiene el gráfico
ax = plt.axes()
# Parrilla:
plt.grid() # Se puede cambiar color, grosor..
# Con plot se puede pilotear información específica
```

```python
# plot en español: Trazar o graficar
import matplotlib.pyplot as plt
x = [1,2,3,5,5,6,7]
y = [1,6,3,4,5,6,7]
plt.plot(x,y) # Función que toma toda la información que se le pase y lo plasma en una gráfica
plt.grid() # Agrega cuadrícula a la gráfica
```

```python
# Plot tiene un tercer elemento llamado format, permite formatear la gráfica
plt.plot(x, y, 'b*:'); # 'o-r' indica que se ploteará un punto de color rojo y una línea de color rojo
# "ro-" r = roja o = puntos del gráfico - = línea
# plt.show() # Muestra la gráfica
# Modificar los ejes
plt.axis([0, 8, 0, 8]) # Modifica los límites de los ejes | Colocar preferiblemente debajo de una a existente porque sola no me mostró nada.
```
