# Estructura Principal de Matplotlib

Antes de avanzar con el curso, quiero que nos tomemos un par de minutos para conocer las partes globales de los gráficos de Matplotlib. 

En la lección anterior vimos cómo construir un gráfico simple, que es este mismo que ahora ves en pantalla.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numeros = [4,2,7,6,3]

plt.plot(numeros);
```

El elemento más grande de nuestros gráficos es el lienzo que los contiene, y se llama **figura**, o `figure`, en inglés. Sería como la pared donde cuelgas tus cuadros, y en este caso, el único cuadro que tiene colgado, es nuestro gráfico.

Puedes acceder al figure de tus gráficos para configurarlo usando todos sus métodos y parámetros, a través de `plt.figure()`.

```python
plt.figure(facecolor="orange")
plt.plot(numeros);
```

El siguiente gran elemento, se llama **axes**, y es el cuadro o área dentro de `figure()`, y es donde realmente se dibujan los gráficos. De hecho, un `figure()` puede tener más de un axes.


El axes incluye no solo el área específica del gráfico, sino también los ejes, las referencias de texto, los valores que se muestran, o (si lo tuviera) su título y las eventuales etiquetas que puede tener junto a cada eje del gráfico.


Para que esto quede claro, te voy a mostrar un ejemplo de un *figure* que tiene más de un gráfico.

```python
plt.subplots(ncols=2, nrows=2)
```

Observa que esta celda me ha devuelto, primero, un par de elementos textuales, y luego un elemento gráfico.

Lo voy a volver a ejecutar, pero esta vez lo voy a asignar a dos variables (`fig` y `axs`), para poder almacenar allí esos elementos.

```python
fig, axs = plt.subplots(ncols=2, nrows=2)
```

Ahora podemos consultar de qué tipo de datos se trata cada uno de ellos.

```python
type(fig)
```

```python
type(axs)
```

Esto es interesante, los axes no son otra cosa que un array. Es decir que lo puedo tratar commo tal. Si es un array entonces puedo indexarlo.

```python
axs[0]
```

```python
axs[0][0]
```

```python
type(axs[0][0])
```

Axes es un objeto de matplotlib, que tiene sus propios métodos y propiedades, los cuales podemos usar para configurar y personalizar todos nuestros gráficos.

Entonces si yo almaceno al primero de estos axes en una variable, luego puedo acceder a sus elementos individuales.

```python
axes1 = axs[0][0]
```

```python
axes1.yaxis
```

Así es como puedes tomar cosas como el título de tu gráfico, las líneas, los ejes, las etiquetas, o lo que sea, y aplicar tus configuraciones.

Por ahora, mi principal objetivo es que comprendas los conceptos de **figura** y **axes**, porque se vienen los gráficos a toda velocidad, y necesitamos entender esta estructura principal para profundizar en ellos sin problemas.
