# Gráficos de Líneas (Line Plots)

Comencemos nuestro recorrido de creación de visualizaciones con la más simple de todas, que son los **gráficos de líneas**.

Ahora que ya sabes cómo están estructurados los gráficos en Matplotlib, quiero agregar algo a lo que hemos visto inicialmente sobre la creación de gráficos, y es que siempre vamos a crear nuestros gráficos de matplotlib creando primero una figura y un axes.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = plt.axes()
```

Aquí hemos creado la pared (la variable `fig`) y el marco vacío del único cuadro que colgaremos en esta pared (`ax`).

La pared es el contenedor de cuadros (uno solo en este caso), y el cuadro es el contenedor de todos los elementos que va a contener nuestro gráfico (los ejes, las etiquetas, las líneas, el título, etc.).

De hecho, los ejes ya se están mostrando, y son los que conforman la caja donde va a desplegarse el gráfico de línea que vamos a crear.

Incluso voy a agregar un elemento más, antes de poner datos aquí, que va a ser el **grid** (o parrilla) para ver unas líneas de guía que nos ayuden a medir nuestros datos.

```python
plt.grid()
```

Si quieres jugar un poco, abre las opciones de `grid()` y revisa sus parámetros y propiedades, para tener un pantallazo de todo lo puedes hacer con este elemento.

Ahora podemos usar au función `plot()` de Matplotlib para *plotear*, o *desplegar*, alguna información específica.

Vamos a crear un conjunto de datos:

```python
x = [1, 2, 3, 5, 5, 6, 7]
y = [1, 6, 3, 4, 5, 6, 7]
```

Veamos la ayuda de `plot()` (usando SHIFT + TAB).

**Plot** en español se podría traducir como *trazar* o *graficar*, entonces cuando en la ayuda dice:

> *Plot y versus x as lines and/or markers.*

...básicamente significa que lo que hace `plot()` es graficar líneas o marcadores en un axes basándose en los valores que se encuentren en dos conjuntos de datos.

Observa en la sintaxis de `plot()`, que el parámetro *args* es un argumento de largo variable, que permite múltiples pares de *x*, *y*.

Para decirlo de un modo más simple, yo lo explicaría así: `plot()` es una función que va a tomar toda la información que le pases, y la va a representar en un gráfico, ya sea con líneas o como marcadores.

Vas a ver que `plot()` es un comando muy flexible, que puede tomar un número arbitrario de argumentos, y pasarlos al gráfico.

Por ejemplo, para plotear, o graficar, los valores de nuestras listas *x* e *y*, lo podemos hacer de esta manera:

```python
plt.plot(x, y);
```

Observa que el grid ya no aparece porque en realidad debemos incluir ese comando en la misma celda donde desplegamos los datos.

```python
plt.plot(x, y)
plt.grid();
```

Lo que se ha representado en este gráfico, es una línea que va siguiendo los diferentes **pares x/y**. El primero es **1/1**, el segundo es **2/6** y así.

##### Formato
Además, por cada par de argumentos *x*/*y* que le pasemos, `plot()` tiene un tercer argumento que es `format`, y que nos permite determinar estilos de líneas y puntos para nuestro gráfico.

Por ejemplo, si le pasamos el argumento `ro-` va a crear **lineas rojas** ('r' por 'red'), con **círculos en los puntos** ('o' representa 'círculo') y con **líneas sólidas** ('-').

```python
plt.plot(x, y, "ro-")
plt.grid();
```

Otro formato posible puede ser `'b*:'`, que significa **color azul** con  **asteriscos en los puntos** y **línea de puntos**

```python
plt.plot(x, y, "b*:")
plt.grid();
```

Por supuesto que hay muchas combinaciones más, y si quieres conocerlas en profundidad, te voy a dejar [este cuaderno extra](Nomenclatura%20para%20Format%20en%20plt.plot().ipynb), con la **nomenclatura completa de los formatos** para tus gráficos.

Otra cosa que podríamos hacer aquí es configurar los **valores máximos y mínimos** que queremos mostrar en el gráfico. Para eso tenemos que convocar a la **configuración de los ejes**.

Voy a hacer de cuenta que no sé cómo se hace esto, para mostrarte la forma de ir descubriendo por ti mismo cómo pensar estas configuraciones.

Supongamos que yo quiero que el área que muestre mi gráfico muestre un poco más de aire en los picos mínimos y máximos para que se vea más limpio. Por lo tanto me gustaría que el **valor mínimo no sea 1, sino 0**, y que el **valor máximo no sea 7, sino 8**, para que haya más espacio en mi gráfico.

En este caso, lo que quiero modificar entonces son los **valores mínimos y máximos de los ejes** ¿cómo logro hacer eso?

Bueno, si lo que quiero modificar son valores que se encuentran en **los ejes**, entonces tengo que buscar esa propiedad (ejes) dentro de matplotlib.

Si escribo `plt.` y presiono **TAB**, puedo revisar la lista de mpetodos disponibles y encontrar uno que se llama `plt.axis` (axis = *ejes* en español).

Esta parece ser la herramienta que busco, pero ¿Cómo puedo estar seguro de que `plt.axis` es la herramienta que busco?

```python
help(plt.axis)
```

Ahora que descubrí que `axis()` es un método que sirve para **configurar algunas propiedades de los ejes**, debo buscar entre sus propiedades para ver si alguna sirve para establecer **valores mínimos y máximos** de sus ejes.

Para nuestra buena suerte, vemos que los primeros parámetros descriptos en la ayuda son justamente `xmin`, `xmax`, `ymin`, e `ymax`, y que según se describe son los **límites de los ejes**. ¡Perfecto!

En la ayuda dice que los podemos pasar como una lista:

```python
plt.plot(x, y, "b*:")
plt.axis([0, 8, 0, 8])
plt.grid();
```
