# Regresión Logística

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split # Sirve para dividir la data
```

```python
df = pd.read_csv("C:\\AmaruAyelen\\proyectos\\PythonDataScienceMachineLearning\\PythonDataScienceMachineLearning\\Files\\Día 11\\Cuadernos para Prácticas\\datos_seguro.csv")
df.head()
```

```python
plt.scatter(df.edad,
            df.compra);
```

```python
# Distribuir los datos para entrenamiento y para prueba en 4 variables
X_entrena, x_aprueba, y_entrena, y_prueba = train_test_split(df[["edad"]],df["compra"],train_size=0.9)
```

```python
X_entrena
```

```python
x_aprueba
```

```python
modelo = LogisticRegression()
```

```python
# Entrenamiento del modelo
modelo.fit(X_entrena, y_entrena)
```

```python
# Evaluación de desempeño
modelo.score(X_entrena, y_entrena)
# En este caso tiene probabilidad del 87% de predecir si una persona comprará o no un seguro.
```

```python
# Evaluar el modelo
datos_nuevos = pd.DataFrame({"edad":[25,25,45,55]})
datos_nuevos
```

```python
# Predecir probabilidades
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades
# En este caso, la columna de la iz es la probabilidad de no compra, y la segunda es la probabilidad de que si
```

```python
prob_compra = probabilidades[:,1] # Así queda el índice 1 de cada uno de los registros
prob_compra
```

```python
plt.scatter(df.edad,
            df.compra);
plt.scatter(datos_nuevos["edad"],
            prob_compra,
            color='red');
```
