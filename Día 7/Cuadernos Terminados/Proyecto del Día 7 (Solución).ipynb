```python
import pandas as pd
```

```python
ruta = 'C:/Users/Win10/Downloads/00 - FEDE/Udemy/Python para Data Science/Día 7/10 - Proyecto del Día 7/Datos_Ventas_Tienda.csv'
df_ventas = pd.read_csv(ruta)
df_ventas
```

```python
ruta2 = 'C:/Users/Win10/Downloads/00 - FEDE/Udemy/Python para Data Science/Día 7/10 - Proyecto del Día 7/Datos_Ventas_Tienda2.csv'
df_ventas2 = pd.read_csv(ruta2)
df_ventas2.head()
```

```python
df = pd.concat([df_ventas, df_ventas2], ignore_index=True)
df.tail()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
df
```

```python
df.info()
```

```python
# Producto más vendido
producto_mas_vendido = df.groupby('Producto')['Cantidad'].sum()
producto_mas_vendido = producto_mas_vendido.sort_values(ascending=False)
producto_mas_vendido.head(1)
```

```python
meses = []
for f in df['Fecha']:
    meses.append(f.month)

df['Meses'] = meses
df
```

```python
ventas_por_mes = df.groupby('Meses')['Total Venta'].sum().sort_values(ascending=False)
ventas_por_mes.head(1)
```

```python
ventas_por_categoria = df.groupby('Producto')['Total Venta'].sum()
ventas_por_categoria
```

```python
df.to_csv('Resumen_ventas.csv')
```
