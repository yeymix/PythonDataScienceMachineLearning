```python
import pandas as pd
```

```python
ruta_airports = "C:\\AmaruAyelen\\proyectos\\PythonDataScienceMachineLearning\\PythonDataScienceMachineLearning\\Files\\Ejercicio grupo 4\\airports.csv"
ruta_routes = "C:\\AmaruAyelen\\proyectos\\PythonDataScienceMachineLearning\\PythonDataScienceMachineLearning\\Files\\Ejercicio grupo 4\\routes.csv"
df_airports = pd.read_csv(ruta_airports)
df_routes = pd.read_csv(ruta_routes)
```

```python
# df_routers_source = df_routes.groupby('Source Airport').size().reset_index(name='Count Routes')
df_routers_source = df_routes.groupby('Source Airport').size().to_frame(name='Count Routes')
df_routers_source_top3 = df_routers_source.sort_values('Count Routes', ascending=False).head(3)
df_routers_source_top3
# print(df_routers_source.iloc[0:3])
# df_routers_airports = df_routers_source_top3.join(df_airports, how='left')
df_routers_airports = df_routers_source_top3.merge(df_airports, left_on='Source Airport', right_on='IATA', how='left')
# df_routers_airports['Date'] = pd.Timestamp.Series(pd.date_range(pd.to_datetime,freq='M2'))
# df_routers_airports['Date'] = pd.Series(pd.date_range('20240101', periods=7, freq='ME'))
start_date = pd.Timestamp.today().normalize() 
df_routers_airports['Date1'] = pd.date_range(start=start_date, periods=7, freq='2MS')
# primera_semana_2024 = pd.Series(pd.date_range('20240101', periods=7, freq='D'))
df_routers_airports[['Country','City','Count Routes','Date1']]

```

```python
# Calcular conteo agrupando por ciudad y país ("Count Routes")​
# df_airports_city_country = df_airports.groupby(['City', 'Country']).count().reset_index()
df_airports_city_country = df_airports.groupby(['City', 'Country']).size().reset_index(name='Count')
print('conteo agrupando por ciudad y país ("Count Routes")​')
print(df_airports_city_country.sort_values('Count', ascending=False))
print("=============================")
# filtro_airports_city_country = df_airports_city_country.loc[df_airports_city_country['City'] == 'Aarhus']
```

```python
# Extraer el top 3 utilizando el método iloc​
df_ordenado = df_airports_city_country.sort_values('Count', ascending=False)
# df_airports_city_country = df_airports.groupby(['City', 'Country']).size().reset_index(name='Count')
df_top_3 = df_ordenado.iloc[0:3]
print("Extraer el top 3 utilizando el método iloc​")
print(df_top_3)
print("=============================")

# top_3 = grouped_sorted.iloc[:3]
```
